```python
import pandas as pd
```

```python
odi=pd.read_csv("./Final ODI Dataset.csv")
```

```python
T20=pd.read_csv("../T20/Final T20 Dataset.csv")
```

```python
test=pd.read_csv("../TEST/Final TEST Dataset.csv")
```

```python
df=pd.concat([odi,test,T20])
```

```python
df[df.Player=="V Kohli"]
```

```python
a=df.groupby(by="Player")[["Matches","Innings","Not_Out","Runs","Ball_Faced","Hundreds","Fiftys","Zeros","Fours","Sixes"]].sum().reset_index()
```

```python
b=df.groupby(by="Player")[["Average","Strike_Rate"]].mean().reset_index()
```

```python
c=df[["Player","Country"]].groupby("Player").first().reset_index()
```

```python
d=df[["Player","Highest_score"]].groupby("Player").max().reset_index()
```

```python
x=a.merge(b)
```

```python
y=x.merge(c)
```

```python
z=y.merge(d)
```

```python
z
```

```python
z[z["Player"]=="V Kohli"]
```

```python
z.to_csv("allplayers.csv")
```
