```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("./Final TEST Dataset.csv")
```

```python
df.drop(columns="Unnamed: 0",inplace=True)
```

```python
df
```

```python
# Top 10 players by Runs
top_10_runs = df.nlargest(10, 'Runs')[['Player', 'Runs', 'Country']]
plt.figure(figsize=(12, 6))
sns.barplot(y='Runs', x='Player', hue='Country', data=top_10_runs,orient="v")
plt.title('Top 10 T20 Players by Runs Scored')
plt.xticks(fontsize=12,rotation=300)

plt.xlabel('Total Runs')
plt.ylabel('Player')
plt.show()
```

```python
df.sort_values(by="Runs",ascending=False)
```

```python
sns.set_style("whitegrid")
sns.barplot(y='Runs', x='Player', hue='Country', data=top_10_runs,orient="v")

plt.title('🏏 Top 10 ODI Players by Runs Scored', fontsize=18, weight='bold')
plt.xlabel('Total Runs', fontsize=14)
plt.ylabel('Player', fontsize=14)
plt.xticks(fontsize=8,rotation=300)
plt.yticks(fontsize=10)
plt.legend(title='Country', fontsize=10, title_fontsize=12)
plt.tight_layout()
```

```python
# Scatter plot: Matches vs. Runs, colored by Country, sized by Career Length
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Matches', y='Runs', hue='Country', sizes=(20, 200), data=df)
plt.title('Matches vs. Runs by Country and Career Length')
plt.xlabel('Matches Played')
plt.ylabel('Total Runs')
plt.tight_layout()
plt.show()
```

```python
# Box plot: Average by Country
plt.figure(figsize=(10, 6))
sns.boxplot(x='Country', y='Average', data=df)
plt.title('Batting Average Distribution by Country')
plt.xlabel('Country')
plt.ylabel('Batting Average')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(12, 8))
corr = df[['Matches', 'Innings', 'Not_Out', 'Runs', 'Average', 'Ball_Faced', 'Strike_Rate', 
           'Hundreds', 'Fiftys', 'Zeros', 'Fours', 'Sixes']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix of ODI Performance Metrics')
plt.show()
```

```python
# Top 20 players by Runs, calculate Zeros per Innings
top_20_runs = df.nlargest(20, 'Runs')[['Player', 'Zeros', 'Country']]
plt.figure(figsize=(12, 6))
sns.barplot(x='Zeros', y='Player', hue='Country', data=top_20_runs)
plt.title('Zeros per Innings for Top 20 Run-Scorers')
plt.xlabel('Zeros per Innings')
plt.ylabel('Player')
plt.show()
```

```python
# Scatter plot: Matches vs. Runs, colored by Country, sized by Career Length
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Matches', y='Runs', hue='Country', sizes=(20, 200), data=df)
plt.title('Matches vs. Runs by Country and Career Length')
plt.xlabel('Matches Played')
plt.ylabel('Total Runs')
plt.tight_layout()
plt.show()
```

```python
sns.set_style("whitegrid")
hundreds=df.sort_values(by="Hundreds",ascending=False).head(10)

sns.barplot(y='Hundreds', x='Player', hue='Country', data=hundreds,orient="v")

plt.title('🏏 Top T20  Players by Runs Scored', fontsize=18, weight='bold')
plt.xlabel('Total Runs', fontsize=14)
plt.ylabel('Player', fontsize=14)
plt.xticks(fontsize=8,rotation=300)
plt.yticks(fontsize=10)
plt.legend(title='Country', fontsize=10, title_fontsize=12)
plt.tight_layout()
```

```python
lables=df[["Country","Hundreds"]].groupby("Country")["Hundreds"].sum()
lables=lables.reset_index()
plt.pie(
    lables.Hundreds,
    
    labels=lables.Country,
    autopct='%1.1f%%',
    shadow=True,
   
)
plt.show()
```

```python
lables=df[["Country","Fiftys"]].groupby("Country")["Fiftys"].sum()
lables=lables.reset_index()
print(lables)
plt.pie(
    lables.Fiftys,
    
    labels=lables.Country,
    autopct='%1.1f%%',
    shadow=True,
   
)
plt.show()
```

```python
# Create line graph
plt.figure(figsize=(12, 6))
sns.lineplot(x='Hundreds', y='Strike_Rate', hue='Country', data=df, marker='o', palette='Set2')
plt.title('Mean Strike Rate by Matches Played Across Countries', fontsize=14)
plt.xlabel('Matches Played (Binned)', fontsize=12)
plt.ylabel('Mean Strike Rate', fontsize=12)
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()

# Save plot
plt.savefig('strike_rate_matches_line_graph.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Fours', y='Runs', hue='Country', size='Fours', sizes=(20, 200), data=df, palette='Set2')
plt.title('Runs vs. Fours Scored by Country', fontsize=14)
plt.xlabel('Number of Fours', fontsize=12)
plt.ylabel('Total Runs', fontsize=12)
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Save plot
plt.savefig('runs_vs_fours_scatter_plot.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python

# Create stacked bar chart
boundary_sums = df.groupby('Country')[['Fours', 'Sixes']].sum().reset_index()

plt.figure(figsize=(10, 6))
boundary_sums.set_index('Country')[['Fours', 'Sixes']].plot(kind='bar', stacked=True, color=['#66c2a5', '#fc8d62'], figsize=(10, 6))
plt.title('Fours and Sixes Contribution by Country', fontsize=14)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Total Boundaries', fontsize=12)
plt.legend(title='Boundary Type')
plt.tight_layout()

# Save plot
plt.savefig('fours_sixes_stacked_bar.png', dpi=300, bbox_inches='tight')
plt.show()
```
