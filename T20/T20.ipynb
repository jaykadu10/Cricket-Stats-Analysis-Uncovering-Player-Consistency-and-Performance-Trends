```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

India = "https://www.espncricinfo.com/records/team/batting-most-runs-career/india-6/twenty20-internationals-3"
Australia ="https://www.espncricinfo.com/records/team/batting-most-runs-career/australia-2/twenty20-internationals-3"
Bangladesh="https://www.espncricinfo.com/records/team/batting-most-runs-career/bangladesh-25/twenty20-internationals-3"
England="https://www.espncricinfo.com/records/team/batting-most-runs-career/england-1/twenty20-internationals-3"
New_zealand="https://www.espncricinfo.com/records/team/batting-most-runs-career/new-zealand-5/twenty20-internationals-3"
Pakistan="https://www.espncricinfo.com/records/team/batting-most-runs-career/pakistan-7/twenty20-internationals-3"
South_africa="https://www.espncricinfo.com/records/team/batting-most-runs-career/south-africa-3/twenty20-internationals-3"
Srilanka="https://www.espncricinfo.com/records/team/batting-most-runs-career/sri-lanka-8/twenty20-internationals-3"
West_indies="https://www.espncricinfo.com/records/team/batting-most-runs-career/west-indies-4/twenty20-internationals-3"
Afghanistan="https://www.espncricinfo.com/records/team/batting-most-runs-career/afghanistan-40/twenty20-internationals-3" 

# headers = {
#     "User-Agent": "Mozilla/5.0 (Linux; Android 11; SAMSUNG SM-G973U) AppleWebKit/537.36 (KHTML, like Gecko) SamsungBrowser/14.2 Chrome/87.0.4280.141 Mobile Safari/537.36"
# }

url=[India,Australia,Bangladesh,England,New_zealand,Pakistan,South_africa,Srilanka,West_indies,Afghanistan]
Country=[]
Player=[]
Span=[]
Matches=[]
Innings=[]
Not_Out=[]
Runs=[]
Highest_score=[]
Average=[]
Ball_Faced=[]
Strike_Rate=[]
Hundreds =[]
Fiftys=[]
Zeros=[]
Fours=[]
Sixes =[]

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}
for x in range(len(url)):
    response = requests.get(url[x], headers=headers)
    print(response)
    soup=BeautifulSoup(response.text)
    x = soup.find_all("td",attrs={"class":"ds-min-w-max"})
    country=soup.find_all("span",attrs={"class":"ds-text-compact-xs ds-block ds-text-typo hover:ds-text-typo-primary"})

    n=0
    while n<len(x):
        Country.append(country[1].text.split()[0])
        Player.append(x[n].text)
        n+=1
        Span.append(x[n].text)
        n+=1
        Matches.append(x[n].text)
        n+=1
        Innings.append(x[n].text)
        n+=1
        Not_Out.append(x[n].text)
        n+=1
        Runs.append(x[n].text)
        n+=1
        Highest_score.append(x[n].text)
        n+=1
        Average.append(x[n].text)
        n+=1
        Ball_Faced.append(x[n].text)
        n+=1
        Strike_Rate.append(x[n].text)
        n+=1
        Hundreds.append(x[n].text)
        n+=1
        Fiftys.append(x[n].text)
        n+=1
        Zeros.append(x[n].text)
        n+=1
        Fours.append(x[n].text)
        n+=1
        Sixes.append(x[n].text)
        n+=1
```

```python
T20={
"Country":Country,
"Player":Player,
"Span":Span,
"Matches":Matches,
"Innings":Innings,
"Not_Out":Not_Out,
"Runs":Runs,
"Highest_score":Highest_score,
"Average":Average,
"Ball_Faced":Ball_Faced,
"Strike_Rate":Strike_Rate,
"Hundreds":Hundreds,
"Fiftys":Fiftys,
"Zeros":Zeros,
"Fours":Fours,
"Sixes":Sixes,
}
```

```python
df=pd.DataFrame(T20)
```
